```python
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
updateResultWorker = pd.read_excel('updateResultWorker.xlsx')
sendMessageWorker = pd.read_excel('SendMessageWorker.xlsx')
insertCouponWorker = pd.read_excel('insertCouponWorker.xlsx')
```

```python
import datetime
start = datetime.datetime(2021, 11, 10, 11,26, 17)
end = datetime.datetime(2021, 11, 10, 11, 26, 39)
temp = start
timeline = []
while temp < end:
    print(temp)
    timeline.append(temp)
    temp = temp + datetime.timedelta(milliseconds=100)
```

```python
sendMessage_timeline = sendMessageWorker['시각']
sendMessage_timeline_100ms = []
for i in sendMessage_timeline:
    strtemp = str(i)
    sendMessage_timeline_100ms.append(strtemp[0:21])
    
count = 0
pointer = ''
x_sendMessage = []
y_sendMessage_count = []
for i in sendMessage_timeline_100ms:
    if i != pointer:
        pointer = i
        x_sendMessage.append(pointer)
        y_sendMessage_count.append(count)
        count = 0
        count += 1
    else:
        count += 1
```

```python
print(x_sendMessage)
```

```python
print(timeline)
```

```python
for i in sendMessage_timeline:
    print(i)
```

```python
timeline_str = []
for i in timeline:
    timeline_str.append(str(i))
```

```python
timeline_sendMessage = []
for i in timeline:
    timeline_sendMessage.append(str(i)[0:21])
```

```python
print(timeline_sendMessage)
```

```python
del y_sendMessage_count[0]
y_sendMessage_count.append(0)
df_sendMessage_before = pd.DataFrame([ x for x in zip(x_sendMessage,y_sendMessage_count)])
```

```python
df_sendMessage_before.head()
```

```python
temp = []
for i in timeline_sendMessage:
    if (df_sendMessage_before[0] == i).any():
        target_count = df_sendMessage_before.loc[df_sendMessage_before[0] == i].iloc[0][1]
        temp.append(target_count)
    else:
        temp.append(0)
```

```python
print(temp) # 타임라인에 맞춘거
```

```python
x_sendMessage = timeline_sendMessage
y_sendMessage_count = temp

print(len(x_sendMessage))
print(len(y_sendMessage_count))
```

```python
insertCoupon_timeline = insertCouponWorker['시각']
insertCoupon_timeline_100ms = []
for i in insertCoupon_timeline:
    strtemp = str(i)
    insertCoupon_timeline_100ms.append(strtemp[0:21])
    
count = 0
pointer = ''
x_insertCoupon = []
y_insertCoupon_count = []
for i in insertCoupon_timeline_100ms:
    if i != pointer:
        pointer = i
        x_insertCoupon.append(pointer)
        y_insertCoupon_count.append(count)
        count = 0
        count += 1
    else:
        count += 1

timeline_insertCoupon = []
for i in timeline:
    timeline_insertCoupon.append(str(i)[0:21])
    
del y_insertCoupon_count[0]
y_insertCoupon_count.append(0)
df_insertCoupon_before = pd.DataFrame([ x for x in zip(x_insertCoupon,y_insertCoupon_count)])


temp2 = []
for i in timeline_insertCoupon:
    if (df_insertCoupon_before[0] == i).any():
        target_count = df_insertCoupon_before.loc[df_insertCoupon_before[0] == i].iloc[0][1]
        temp2.append(target_count)
    else:
        temp2.append(0)
        
x_insertCoupon = timeline_insertCoupon
y_insertCoupon_count = temp2

print(len(x_insertCoupon))
print(len(y_insertCoupon_count))
```

```python
updateResult_timeline = updateResultWorker['시각']
updateResult_timeline_100ms = []
for i in updateResult_timeline:
    strtemp = str(i)
    updateResult_timeline_100ms.append(strtemp[0:21])
    
count = 0
pointer = ''
x_updateResult = []
y_updateResult_count = []
for i in updateResult_timeline_100ms:
    if i != pointer:
        pointer = i
        x_updateResult.append(pointer)
        y_updateResult_count.append(count)
        count = 0
        count += 1
    else:
        count += 1

timeline_updateResult = []
for i in timeline:
    timeline_updateResult.append(str(i)[0:21])
    
del y_updateResult_count[0]
y_updateResult_count.append(0)
df_updateResult_before = pd.DataFrame([ x for x in zip(x_updateResult,y_updateResult_count)])


temp3 = []
for i in timeline_updateResult:
    if (df_updateResult_before[0] == i).any():
        target_count = df_updateResult_before.loc[df_updateResult_before[0] == i].iloc[0][1]
        temp3.append(target_count)
    else:
        temp3.append(0)
        
x_updateResult = timeline_updateResult
y_updateResult_count = temp3

print(len(x_updateResult))
print(len(y_updateResult_count))
```

```python
fig_all = plt.gcf()
fig_all.set_size_inches(300, 20)

plt.grid(True, axis='x')
plt.xticks(fontsize=16, rotation='vertical')
plt.yticks(fontsize=33)
plt.bar(timeline_str, temp) #sendMessage
plt.bar(timeline_str, temp2) #insertCoupon
plt.bar(timeline_str, temp3) #updateResult
plt.savefig('triple_worker.png', dpi=100)
```
